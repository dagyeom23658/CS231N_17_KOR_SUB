```python
f = open("/content/drive/MyDrive/NER/말뭉치 - 형태소_개체명/00002_NER.txt", 'r')
input_list = f.readlines()
f.close()
print("--예시--")
print(input_list[0:10], "\n 총 문장 수 : ", len(input_list))
```

```python
li = []
li2 = []
for i in range(712):
    if input_list[i].startswith("##"):  # 만약 ## 로 시작한다면
        li2.append(li)  
        li = []         # 빈리스트에
        li.append(i)   # 숫자만 추가해준다. 
    else:
        li.append(input_list[i])  # ##으로 시작하지 않는 문자는 글자 모두를 추가해준다. 
li2.append(li)
```

```python
li2_drop = []
for l in li2:
    if len(l) ==0 or len(l)==1 or l == ["\n"]:  # 빈 리스트거나, li개수가 1개인거 빼고 li2_drop에 담기
        continue
    else:
        li2_drop.append(l[1:])
li2_drop
```

```python
li_0 = []
li_1 = []
li_2 = []
li_3 = []
li_4 = []

for i in range(len(li2_drop)):
    print(i+1)
    for j in li2_drop[i]:
        m = j.split("\t")
        if m[0] == '_' or m[0]== '\n':
            continue

        else:
            li_0.append(f'Sentence: {i+1}')
            li_1.append(m[0])
            li_2.append(m[1])
            li_3.append(m[2])
            li_4.append(m[3])

# https://rfriend.tistory.com/504
```

```python
df_tmp = pd.DataFrame({'Sentence #': li_0,
                        'tokernize1': li_1, 
                        'tokernize2': li_2, 
                        '품사_tag': li_3,
                        '개체명_tag': li_4 })
df_tmp.tail()
```

```python
df_tmp['개체명_tag'] = df_tmp['개체명_tag'].apply(lambda x: x.replace("\n",""))
df_tmp
```
